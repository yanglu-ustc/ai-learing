B. 【  】双向RNN，因为这允许反向传播计算中有更精确的梯度。

C. 【  】单向RNN，因为$y^{<t>}$的值仅依赖于$x^{<1>},…,x^{<t>}$，而不依赖于$
x^{<t+1>},…,x^{<365>}$
