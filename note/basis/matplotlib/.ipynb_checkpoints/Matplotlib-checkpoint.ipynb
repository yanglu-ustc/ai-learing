# Matplotlib
## mat - matrix - 矩阵
## plot - 画图
## lib - library - 库
+ 画图，数据可视化
+ 帮助理解数据，方便选择更合适的分析方法

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 创建画布
plt.figure()
# 创建参数，将数据描起来
plt.plot([1,0,9],[4,5,6])
# 展示
plt.show()
```

# 折线图(plot)与基础画图功能

### 折线图的绘制与显示

```python
# 创建画布
plt.figure()
# 绘制图像
plt.plot([1,2,3,4,5,6,7],[17,17,18,15,11,11,13])
# 显示图像
plt.show()
```

```python
# 设置画布属性与图片保存
# plt.figure(figsize=(),dpi=)
# figsize:指定图的长宽
# dpi:图像的清晰度
# 返回fig对象
# plt.savefig(path)保存图片
plt.figure(figsize=(20,8),dpi=80)
plt.plot([1,2,3,4,5,6,7],[17,17,18,15,11,11,13])
plt.savefig("test_plot.png")
# 注意：plt.show()会释放figure资源，如果在显示图像之后保存就会导致只能保存空文件
plt.show()
```

### 完善原始折线图（显示辅助层）

#### 准备数据并画出初始折线图

```python
import random
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
plt.show()
```

```python
y_shanghai
```

#### 添加自定义x,y刻度

```python
# plt.xticks(x,**kwargs) x：要显示的刻度
# plt.yticks(y,**kwargs) y：要显示的刻度
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
# y：要显示的刻度
plt.yticks(range(40))
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
# 设置段的长度
plt.yticks(range(0,40,5))
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
# 也可以这么做
plt.xticks(x[::5])
plt.yticks(range(40)[::5])
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
# 准备x的刻度说明
x_label = [f"11:{i}" for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(40)[::5])
plt.show()
```

#### 添加网格显示

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
x_label = [f"11:{i}" for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(40)[::5])

# 添加显示网格
# linestyle：网格方式
# alpha：透明度
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
```

#### 添加描述信息

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai)
x_label = [f"11:{i}" for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(40)[::5])
plt.grid(True,linestyle="--",alpha=0.5)

# 添加描述信息
plt.xlabel("time")
plt.ylabel("temperature")
plt.title("shanghai's temperature one day")
plt.show()
```

### 完善原始折线图（图像层）

```python
# 多次plot
y_beiging = [random.uniform(1,3) for i in x]

plt.figure(figsize=(20,8),dpi=80)
# 设置颜色，线条风格，设置名称标签
plt.plot(x,y_shanghai,color="r",linestyle="--",label="shanghai")
plt.plot(x,y_beiging,color="b",label="beijing")

# 显示图例：标签
# 可以使用loc设置位置
plt.legend()

x_label = [f"11:{i}" for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(40)[::5])
plt.grid(True,linestyle="--",alpha=0.5)

# 添加描述信息
plt.xlabel("time")
plt.ylabel("temperature")
plt.title("shanghai and beijing's temperature one day")
plt.savefig("test_plot.png")
plt.show()
```

### 多个坐标系显示-plt.subplots()：面向对象的画图方法

```python
# matplotlib.pyplot.subplots(nrows=1,ncols=1,**fig_kw)创建一个带有多个axes（坐标系/绘图区）的图
# return：返回的是两个值，一个是画布，一个是绘画区
# figure,axes=plt.subplots(nrows=1,ncols=1,**fig_kw)
# axes[0]/axes[1]代表两个区域
# 一般的方法名会变成set_xxx()的形式
```

```python
# 多次plot
y_beiging = [random.uniform(1,3) for i in x]

# plt.figure(figsize=(20,8),dpi=80)
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="shanghai")
axes[1].plot(x,y_beiging,color="b",label="beijing")

# legend()不需要set_xxx()
axes[0].legend()
axes[1].legend()

x_label = [f"11:{i}" for i in x]
axes[0].set_xticks(x[::5],x_label[::5])
axes[0].set_yticks(range(40)[::5])
axes[1].set_xticks(x[::5],x_label[::5])
axes[1].set_yticks(range(40)[::5])

# grid()不需要set_xxx()
axes[0].grid(True,linestyle="--",alpha=0.5)
axes[1].grid(True,linestyle="--",alpha=0.5)

# 添加描述信息
axes[0].set_xlabel("time")
axes[0].set_ylabel("temperature")
axes[0].set_title("shanghai s temperature one day")
axes[1].set_xlabel("time")
axes[1].set_ylabel("temperature")
axes[1].set_title("beijing's temperature one day")
plt.show()
```

```python
# 多次plot
y_beiging = [random.uniform(1,3) for i in x]

# plt.figure(figsize=(20,8),dpi=80)
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="shanghai")
axes[1].plot(x,y_beiging,color="b",label="beijing")

# legend()不需要set_xxx()
axes[0].legend()
axes[1].legend()

# 有的时候会出问题，可以使用这种方式设置坐标
x_label = [f"11:{i}" for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(40)[::5])

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(40)[::5])

axes[0].grid(True,linestyle="--",alpha=0.5)
axes[1].grid(True,linestyle="--",alpha=0.5)

# 添加描述信息
axes[0].set_xlabel("time")
axes[0].set_ylabel("temperature")
axes[0].set_title("shanghai s temperature one day")
axes[1].set_xlabel("time")
axes[1].set_ylabel("temperature")
axes[1].set_title("beijing's temperature one day")
plt.show()
```

```python
# 扩展：plt.plot()可以画出各种函数图像
# 设置的点足够密集，就可以产生函数图像
import numpy as np

# 准备数据
x = np.linspace(-6,6,1000)
y = np.sin(x)

# 创建画布
plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# 散点图(scatter)

```python
# 用于观察数据，查找规律
# 判断变量之间是否存在数量关联趋势，展示离群点（分布规律）
x = range(60)
y = [random.uniform(15,18) for i in x]
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)
axes[0].plot(x,y)
# 散点图
axes[1].scatter(x,y)
plt.show()
```

# 柱状图(bar)

```python
# 绘制连离散的数据，一眼能够看出各个数据的大小，比较数据之间的差别（统计/对比）
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)
axes[0].plot(x,y)
# 散点图
axes[1].bar(x,y)
plt.show()
```

```python
# 对比
movie_names = ["Dune: Part Two","Kung Fu Panda 4","Godzilla x Kong: The New Empire",
               "Bob Marley: One Love","The Beekeeper","Ghostbusters: Frozen Empire","Mean Girls","Madame Web","Argylle","Exhuma"]
tickets=[631278969,
         351181555,
    	214305986,
    	175922424,
    	152720535,
    	113121375,
    	104404248,
    	99767157,
    	95340101,
    	88221322]

plt.figure(figsize=(20,8),dpi=80)
x_tickes = range(len(movie_names))
# 自动对颜色进行循环！！！
plt.bar(x_tickes,tickets,color=['b','r','g','y','c','m'])
plt.xticks(x_tickes,movie_names)
plt.title("the movies' tickets around the world")

plt.grid(linestyle="-.",alpha=0.8)

plt.show()
```

# 直方图(histog)

```python
# 由一系列高度不等的纵向条纹或线段表示数据的分布情况。一般用横轴表示数据范围，纵轴表示分布情况
# 绘制连续性的数据展示一组或者多组数据的分布情况
# 直方图的柱子宽度是不一样的，这个宽度是有意义的
time = [random.randint(70,140) for i in range(250)]
plt.figure(figsize=(20,8),dpi=80)

distance = 2
group_num=int((max(time)-min(time))/distance)
# bins：组数
plt.hist(time,bins=group_num)

# 修改x轴刻度
plt.xticks(range(min(time),max(time),distance))

plt.title("the ...................... i don't know")
plt.grid(linestyle="-.",alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=80)

# 属性density=True：表示使用频率分布直方图
plt.hist(time,bins=group_num,density=True)

plt.xticks(range(min(time),max(time),distance))
plt.title("the ...................... i don't know")
plt.grid(linestyle="-.",alpha=0.5)
plt.show()

# 频数除以样本总个数为频率，纵轴表示频率除以组距！！！！！
```

# 饼图(pie)

```python
# 用于表示不同类别的占比情况，通过弧度大小来对比各种分类
# 分类数据的占比情况
movie_names = ["Dune: Part Two","Kung Fu Panda 4","Godzilla x Kong: The New Empire",
               "Bob Marley: One Love","The Beekeeper","Ghostbusters: Frozen Empire","Mean Girls","Madame Web","Argylle","Exhuma"]
tickets=[631278969,
         351181555,
    	214305986,
    	175922424,
    	152720535,
    	113121375,
    	104404248,
    	99767157,
    	95340101,
    	88221322]

plt.figure(figsize=(20,8),dpi=80)
plt.pie(tickets,labels=movie_names,autopct="%1.2f%%",colors=['b','r','g','y','c','m','k'])
plt.legend()

# 保证正圆
plt.axis('equal')
plt.show()
```
