```python
import numpy as np
from matplotlib import pyplot as plt

# 一元一次函数图像
z = np.arange(-10, 10, 0.01)
a = 1 / (1 + np.power(np.e, -z))
plt.xlabel('z')
plt.ylabel('a')
plt.title("sigmoid-function")
plt.plot(z, a)
plt.grid()
plt.show()
```

```python
a = np.tanh(z)
plt.xlabel('z')
plt.ylabel('a')
plt.title("tanh-function")
plt.plot(z, a)
plt.grid()
plt.show()
```

```python
a = np.maximum(np.zeros(2000),z)
plt.xlabel('z')
plt.ylabel('a')
plt.title("ReLU-function")
plt.plot(z, a)
plt.grid()
plt.show()
```

```python
a = np.maximum(0.01*z,z)
plt.xlabel('z')
plt.ylabel('a')
plt.title("leaky ReLU-function")
plt.plot(z, a)
plt.grid()
plt.show()
```
