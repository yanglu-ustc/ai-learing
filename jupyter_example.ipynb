#### 快捷键：
+ 运行代码：shift+enter —— 跳转下一单元
+ 运行代码：ctrl+enter —— 留在本单元
#### cell操作
+ enter：编辑模式
+ esc：命令模式
    + Y：切换到code模式
    + M：切换到markdown模式
    + A：上面添加一个cell
    + B：下面添加一个cell
    + 双击D：删除当前的cell
    + Z：回退
    + L：为当前cell加上行号

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
x = range(100)
x_label = ["" for i in x]
plt.xticks(x[::1],x_label[::1])
y = range(100)
y_label = ["" for i in y]
plt.yticks(y[::1],y_label[::1])
plt.grid(True,linestyle="-",alpha=0.5)

plt.savefig("坐标纸.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(17,8))
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(1))    #设置x轴主刻度
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))   #设置x轴次刻度
ax.yaxis.set_major_locator(plt.MultipleLocator(1))    #设置x轴次刻度
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))    #设置x轴次刻度
ax.grid(which='major',axis="both",linewidth=0.75,linestyle='-',color='r')
ax.grid(which='minor',axis="both",linewidth=0.25,linestyle='-',color='r')
x=np.linspace(0,3*np.pi,50)#-1到1中画50个点
y=0.000001*np.sin(x)
plt.plot(x,y)
plt.margins(x=0)
plt.margins(y=0)

plt.savefig("my.png")
plt.show()

```
